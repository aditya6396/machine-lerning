```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as mp
import numpy as np



```

```python

df = pd.read_csv("weather_data_cities.csv")


# draw lineplot
df


```

```python
sns.lineplot(x=df.city,y=df.temperature,)
sns.set_style("darkgrid")
```

```python
mp.figure(figsize =(10,5))
mp.subplot(221)
mp.plot(df['city'],df['temperature'],color='red',)
mp.subplot(222)
sns.lineplot(x=df.city,y=df.temperature)
mp.subplot(223)
mp.plot(df['windspeed'],df['event'],color='red',)
mp.subplot(224)
sns.lineplot(x=df.windspeed,y=df.event)

mp.show()
```

```python
x1 = ["city", "temperature", "wind", "speed"]
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])
mp.bar(x1, y1, color='r')

mp.bar(x1, y2, bottom=y1, color='b')

mp.bar(x1, y3, bottom=y1+y2, color='y')

mp.bar(x1, y4, bottom=y1+y2+y3, color='g')

mp.xlabel('x-axis', fontweight ='bold', fontsize = 15)


mp.show()



```

```python
#10
#WED-27thAPR-Day10
#Q. Formulate PDF and CDF functions for normal distribution; pass the values of(x,mu,sigma);
#check whether x and mu are in the form  of std. normal distribution or not; if not,
#convert it into std. normal form.




import math
import statistics as st
#Function to find pdf of Normal distribution
def pdf_normal(x):
    mn=st.mean(x)
    std=st.stdev(x)
    z=[]
    for i in x:
        z.append((i-mn)/std)
    pdflst=[]
    pdf=0
    for i in z:
        pdf+=(math.exp(((-1)*(i**2))/2))/(std*(math.sqrt(2*(math.pi))))
        pdflst.append(pdf)
    return pdflst

xrange=range(0,10,1)
pdf_normal(xrange)

#create plot for PDF
import matplotlib.pyplot as plt
plt.bar(xrange, pdf_normal(xrange))
plt.xlabel("Observations")
plt.ylabel("Probability Density Function")
plt.title("Probability Density Function")
plt.show()

#Function to find cdf of Normal Distribution
def cdf_normal(x):
    mn=st.mean(x)
    std=st.stdev(x)
    z=[]
    for i in x:
        z.append(math.erf((i-mn)/(math.sqrt(2)*std)))
    cdflst=[]
    cdf=0
    for i in z:
        cdf+=(1+i)/2
        cdflst.append(cdf)
    return cdflst

cdf_normal(xrange)

#create plot for CDF
import matplotlib.pyplot as plt
plt.bar(xrange, cdf_normal(xrange))
plt.xlabel("Observations")
plt.ylabel("Cumulative Density Function")
plt.title("Cumulative Density Function")
plt.show()

from scipy import special
#Find x from z score
def find_x(cdf,mn,std):
    x=[]
    for i in cdf:
        calx=(math.sqrt(2)*std*(special.erfinv((2*i)-1)))+mn
        x.append(calx)
    return x

mn=st.mean(xrange)
std=st.stdev(xrange)

find_x(cdf_normal(xrange),mn,std)
```

Q. Find MAE(mean absolute error), MSSE(mean sum of squared error) and RMSE(root mean sum of squared error). Take two random arrays of size 50 as actual and predicted values.

```python
import numpy as np
```

y_actual = np.random.randint(0,10,50)
y_predicted = np.random.randint(0,10,50)

```python
error = x_y = y_actual - y_predicted
```

```python
def MAE(x_y):
    mae = sum(abs(x_y)/len(x_y))
    return "mean absolute error :- ",mae
```

```python
MAE(error)
```

```python
def SSE(x_y):
    sse = sum((x_y)**2)
    return "sum of squared error :-",sse
```

```python
SSE(error)
```

```python
def RMSE(x_y):
    rmse = (sum((x_y)**2))**0.5
    return "root mean squared error :-",rmse
```

```python
RMSE(error)
```

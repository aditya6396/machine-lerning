3) Use “bill_authentication.csv” dataset. Your task is to predict whether a bank note is authentic or fake depending upon the four different attributes of the image of the note. The attributes are variance of wavelet transformed image, curtosis of the image, entropy and skewness of the image.

```python
import pandas as pd 
import numpy as np
 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("bill_authentication.csv")
df
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
class1= df.groupby('Class').Class.count()
class1
```

```python
#plot single column by count its value by using count plot
plt.title("Class")
sns.countplot(df["Class"])
plt.show()
```

```python
target = df.Class
df.drop(['Class'], axis=1, inplace = True)
```

```python
y = target
X = df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, shuffle=True, random_state=10)
```

```python
from sklearn.svm import SVC

model = SVC(kernel='linear').fit(X_train, y_train)
print(model.score(X_train, y_train))
```

```python
y_pred = model.predict(X_val)
print(model.score(X_val, y_val))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_val,y_pred))
print(classification_report(y_val,y_pred))
```

```python
# K-nearest neighbours and classification report
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier()
neigh.fit(X_train, y_train)
neigh.score(X_train, y_train)
```

```python
y_pred = neigh.predict(X_val)
print(neigh.score(X_val, y_val))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_val,y_pred))
print(classification_report(y_val,y_pred))
```

Comparison of accuracy of KNN and SVC models
As shown in the classification report and confusion matrix, KNN model has a 100% accuracy for the test/validation set whereas SVC model has an accuracy of 99%. This could mean the the KNN model fits the dataset better than the SVC model where it's possible to accurately predict whether the bank note is forged by analysis the mode of the 5 nearest neighbour. However, the accuracy for SVC model is also considered respectable at 99%.
